```python
import numpy as np
import networkx as nx
from scipy import spatial

import random
from itertools import cycle

from matplotlib import animation
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
```

```python
class RRT_Bidirectional_Connect():
    def __init__(self, robot, start, goal, tolerance, delta = 1/180.0, goalBias = 0.05):

        self.tree1 = nx.DiGraph()
        self.tree2 = nx.DiGraph()

        self.n_iters = 0
        self.n_extensions = 0

        self.robot = robot
        self.start = start
        self.goal  = goal
        self.tolerance  = tolerance
        self.goal_found  = False
        self.delta = delta
        self.goalBias = goalBias
        
        self.next_node_id = 0
        self.start_node_id = self.get_node_id()
        self.tree1.add_node(self.start_node_id, attr_dict={'state':self.start})
        
        self.goal_node_id = self.get_node_id()
        self.tree2.add_node(self.goal_node_id, attr_dict={'state':self.goal})
        
    def get_node_id(self):
        _id = self.next_node_id
        self.next_node_id += 1
        return _id
    
    def nearest_neighbor(self, tree, state):
        node_so_far = None
        distance_so_far = None
        for this_node in tree.nodes_iter():
            this_distance = self.distance(tree.node[this_node]['state'],state) 
            if(distance_so_far is None):
                distance_so_far = this_distance
                node_so_far = this_node
            elif(distance_so_far > this_distance):
                node_so_far = this_node
                distance_so_far = this_distance
        return (node_so_far,distance_so_far)
    
    def search(self,iters=500):
        iters = int(iters)
        trees = cycle([[self.tree1, self.tree2],[self.tree2, self.tree1]])
        biased = False
        
        for i in xrange(iters):
            self.n_iters += 1
            treeA, treeB = trees.next()
            
            if (np.random.rand() <= self.goalBias):
                q_rand = self.sample_tree(treeA)
                biased = True
            else:
                q_rand = self.sample()
                biased = False
            
            self.extend(q_rand, treeA, treeB, biased)
            
            if self.goal_found:
                return True
        return False
    
    def extend(self, q_rand, treeA, treeB, biased):
        self.n_extensions += 1

        if(biased):
            q_rand_id = q_rand
            q_rand = treeA.node[q_rand]['state']
            
        q_nearest_id, q_nearest_distance  = self.nearest_neighbor(treeB, q_rand)
        q_new_id = self.extend_from(treeB, q_nearest_id, q_nearest_distance, q_rand)
        
        if(biased):
            if(q_new_id is not None):
#                 print('q_rand',q_rand)
#                 print('q_new_id',q_new_id)
#                 print("treeB.node[q_new_id]['state']",treeB.node[q_new_id]['state'])
                if(np.array_equal(q_rand, treeB.node[q_new_id]['state'])):
                    treeB.add_node(q_rand_id,attr_dict={'state': q_rand})
                    treeB.add_edge(q_new_id, q_rand_id)
            
    
    def extend_from(self, tree, q_nearest_id, q_nearest_distance, q_rand):
        q_new_id = self.extend_lin(tree, q_nearest_id, q_nearest_distance, q_rand)
#         if(q_new_id is None):
#             print('WTF?')
        return q_new_id
    
    def extend_lin(self, tree, q_nearest_id, q_nearest_distance, q_rand):
        num = np.ceil(q_nearest_distance/self.delta)
        
        if(num < 2):
            collision_free = self.collision_free(q_rand)   
            if collision_free:
                q_new_id = self.get_node_id()
                tree.add_node(q_new_id,attr_dict={'state': q_rand})
                tree.add_edge(q_nearest_id,q_new_id)
    #             print tree.node[q_new_id]
                self.check_goal(tree, q_new_id)
                return q_new_id
        
        q_near = tree.node[q_nearest_id]['state']
        
        theta1 = np.linspace(q_near[0], q_rand[0], num)[1:]
        theta2 = np.linspace(q_near[1], q_rand[1], num)[1:]
        theta3 = np.linspace(q_near[2], q_rand[2], num)[1:]
        q_s = np.vstack([theta1,theta2,theta3]).T
        
        q_prev_id = q_nearest_id
        for i in range(len(q_s)):
            if(i >= 1):
                q = q_s[i,:]
#                 print('q', q)
#                 print('q_prev_id', q_prev_id)
                collision_free = self.collision_free(q)   
                if collision_free:
                    q_new_id = self.get_node_id()
                    tree.add_node(q_new_id,attr_dict={'state': q})
                    tree.add_edge(q_prev_id,q_new_id)
                    self.check_goal(tree, q_new_id)
                    q_prev_id = q_new_id
                else:
                    return q_prev_id

    def steer(self, tree, q_nearest_id, q_rand):       
        q_near = tree.node[q_nearest_id]['state']
        direction = q_rand - q_near
        unit_direction = direction / np.linalg.norm(direction)
        action = unit_direction * self.delta
        q_new = q_near + action
        return q_new, action
    
    def sample_tree(self, tree):       
        nodes = tree.nodes()
        q_rand_id = random.choice(nodes)
#         q_rand = tree.node[q_rand_id]['state']
        return q_rand_id
        
    def best_solution_(self, tree, state):

        path = []        
        parent = self.nearest_neighbor(tree, state)[0]
        path.append(parent)
        
        while(parent != self.start_node_id): #until you get at the root
            parent = tree.predecessors(parent)                
            assert len(parent)==1
            parent = parent[0]
            path.append(parent)            
        return path[::-1] #reverse
        
    def best_solution(self):
        self.tree = nx.compose(self.tree1,self.tree2)
        graph = nx.compose(self.tree1,self.tree2.reverse())
        path = nx.shortest_path(graph, source=0, target=1)
        return path
    
    
    def check_goal(self, tree, node_id):
        nodesA = self.tree1.nodes()
        nodesB = self.tree2.nodes()
#         print('nodesA',nodesA)
#         print('nodesB',nodesB)
#         print('bool(set(nodesA) & set(nodesB))',bool(set(nodesA) & set(nodesB)))
        self.goal_found = bool(set(nodesA) & set(nodesB))
        
    def collision_free(self,q):
        """
        collision_check(state) = True/False
        """
        return self.robot.collision_free(q)
    
    def sample(self):
        """
        sample() returns point in state space
        """
        return np.random.uniform(-np.pi, np.pi, size=3)
    
    def distance(self,u,v):
        """
        distance(from_node,to_point)
        """
        return spatial.distance.euclidean(u, v)
```
