```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation

import matplotlib.pyplot as plt

import numpy as np
```

```python
class UpdateTree(object):
    def __init__(self, ax, rrt):
        
        self.ax  = ax
        self.rrt = rrt
        self.nodes = self.rrt.tree1.nodes()
        self.lines = [self.ax.plot([0,0],[0,0],[0,0])[0]]
    
    def draw(self,i):
        node_id = self.nodes[i]
        
        node = self.rrt.tree1.node[node_id]
        q_new = node['state']
        
        parent = self.rrt.tree1.predecessors(node_id)                
        assert len(parent) == 1
        parent = parent[0]
        q_near = self.rrt.tree1.node[parent]['state']
        
        line = np.array([q_new,q_near]).T
        line = self.ax.plot(line[0], line[1], line[2])[0]
        
        self.lines.append(line)
        return self.lines
        

    def init(self):
        return self.lines

    def __call__(self, i):
        return self.draw(i)

def animateTree(rrt, title):
    my_dpi = 80
    fig = plt.figure(figsize=(1080/my_dpi, 720/my_dpi), dpi=my_dpi)
    ax  = p3.Axes3D(fig, aspect='equal')

    ax.set_xlim3d(-np.pi,np.pi)
    ax.set_xlabel('$\Theta1$ rad')
    ax.set_ylim3d(-np.pi,np.pi)
    ax.set_ylabel('$\Theta2$ rad')
    ax.set_zlim3d(-np.pi,np.pi)
    ax.set_zlabel('$\Theta3$ rad')
    ax.set_title(title)

    ud = UpdateTree(ax=ax, rrt=rrt)
    anim = animation.FuncAnimation(fig, ud, frames=len(ud.nodes), init_func=ud.init, interval=1, blit=False)
    # plt.show()
    # mywriter = animation.FFMpegWriter()
    # anim.save('RRT_Baseline.mp4', writer=mywriter, fps=300, dpi=my_dpi)
    return anim
```

```python
class UpdateArm(object):
    def __init__(self, ax, rrt, path):
        
        self.ax  = ax
        self.rrt = rrt
        self.rrt.robot.ax = ax
        self.path = path
    
    def draw(self,i):
        state = self.path[i]
        self.rrt.robot.draw_robot(state)
    
    def __call__(self, i):
        return self.draw(i)

def animateArm(rrt, title, path):
    my_dpi = 80
    fig = plt.figure(figsize=(1080/my_dpi, 720/my_dpi), dpi=my_dpi)
    ax = plt.axes(aspect='equal')
    
    ax.set_title(title)

    ud = UpdateArm(ax=ax, rrt=rrt, path=path)
    anim = animation.FuncAnimation(fig, ud, frames=len(ud.path), interval=1, blit=False)
    return anim
```
