```python
class RRT_Baseline():
    def __init__(self, robot, start, goal, tolerance, delta = 1/180.0):

        self.tree1 = nx.DiGraph()

        self.n_iters = 0
        self.n_extensions = 0

        self.robot = robot
        self.start = start
        self.goal  = goal
        self.tolerance  = tolerance
        self.goal_found  = False
        self.delta = delta
        
        self.next_node_id = 0
        self.start_node_id = self.next_node_id
        self.tree1.add_node(self.start_node_id, attr_dict={'state':self.start})
        
    def get_node_id(self):
        _id = self.next_node_id
        self.next_node_id += 1
        return _id
    
    def nearest_neighbor(self, tree, state):
        node_so_far = None
        distance_so_far = None
        for this_node in tree.nodes_iter():
            this_distance = self.distance(tree.node[this_node]['state'],state) 
            if(distance_so_far is None):
                distance_so_far = this_distance
                node_so_far = this_node
            elif(distance_so_far > this_distance):
                node_so_far = this_node
                distance_so_far = this_distance
        return (node_so_far,distance_so_far)
    
    def search(self,iters=500):
        iters = int(iters)
        for i in xrange(iters):
            self.n_iters += 1
            q_rand = self.sample()
            self.extend(q_rand)
            if self.goal_found:
                return True
        return False
    
    def extend(self,q_rand):
        self.n_extensions += 1
        tree = self.tree1

        q_nearest_id, nearest_distance  = self.nearest_neighbor(tree, q_rand)
        x_new_id = self.extend_from(tree, q_nearest_id, q_rand)
    
    def extend_from(self, tree, q_nearest_id, q_rand):

        q_new, action = self.steer(tree, q_nearest_id, q_rand)
        
        collision_free = self.collision_free(q_new)        
        if collision_free:
            q_new_id = self.get_node_id()
            tree.add_node(q_new_id,attr_dict={'state': q_new})
            tree.add_edge(q_nearest_id,q_new_id)
            self.check_goal(tree, q_new_id)
            return q_new_id
        
        return None

    def steer(self, tree, q_nearest_id, q_rand):       
        q_near = tree.node[q_nearest_id]['state']
        direction = q_rand - q_near
        unit_direction = direction / np.linalg.norm(direction)
        action = unit_direction * self.delta
        q_new = q_near + action
        return q_new, action
        
    def best_solution_(self, tree, state):

        path = []        
        parent = self.nearest_neighbor(tree, state)[0]
        path.append(parent)
        
        while(parent != self.start_node_id): #until you get at the root
            parent = tree.predecessors(parent)                
            assert len(parent)==1
            parent = parent[0]
            path.append(parent)            
        return path[::-1] #reverse
        
    def best_solution(self, tree, state):
        goal_id = self.nearest_neighbor(tree, state)[0]
        graph = tree.reverse()
        path = nx.shortest_path(graph, source=goal_id, target=self.start_node_id)
        return path[::-1]
    
    
    def check_goal(self, tree, node_id):
        distance = self.distance(self.goal, tree.node[node_id]['state'])
        self.goal_found = distance <= self.tolerance
        
    def collision_free(self,q):
        """
        collision_check(state) = True/False
        """
        return self.robot.collision_free(q)
    
    def sample(self):
        """
        sample() returns point in state space
        """
        return np.random.uniform(0, 2*np.pi, size=3)
    
    def distance(self,u,v):
        """
        distance(from_node,to_point)
        """
        return spatial.distance.euclidean(u, v)
```

```python
"""
A simple example of an animated plot... In 3D!
"""
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d as p3
import matplotlib.animation as animation

def Gen_RandLine(length, dims=2) :
    """
    Create a line using a random walk algorithm

    length is the number of points for the line.
    dims is the number of dimensions the line has.
    """
    lineData = np.empty((dims, length))
    lineData[:, 0] = np.random.rand(dims)
    for index in range(1, length) :
        # scaling the random numbers by 0.1 so
        # movement is small compared to position.
        # subtraction by 0.5 is to change the range to [-0.5, 0.5]
        # to allow a line to move backwards.
        step = ((np.random.rand(dims) - 0.5) * 0.1)
        lineData[:, index] = lineData[:, index-1] + step

    return lineData

def update_lines(num, dataLines, lines) :
    for line, data in zip(lines, dataLines) :
        # NOTE: there is no .set_data() for 3 dim data...
        line.set_data(data[0:2, :num])
        line.set_3d_properties(data[2,:num])
    return [lines]

# Attaching 3D axis to the figure
fig = plt.figure()
ax = p3.Axes3D(fig, aspect='equal')

# Fifty lines of random 3-D lines
data = [Gen_RandLine(25, 3) for index in range(50)]

# Creating fifty line objects.
# NOTE: Can't pass empty arrays into 3d version of plot()
lines = [ax.plot(dat[0, 0:1], dat[1, 0:1], dat[2, 0:1])[0] for dat in data]

# Setting the axes properties
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('RRT')

# Creating the Animation object
line_ani = animation.FuncAnimation(fig, update_lines, 30, fargs=(data, lines), interval=50, blit=False)

plt.show()
```
